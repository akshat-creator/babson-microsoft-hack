```python
from streetview import *
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
```

```python
start = (42.356280, -71.062290)  # [latitude, longitude]
goal = (42.35155, -71.05818)  # [latitude, longitude]

all_imgs = get_path_imgs(start=start, end=goal, interval=1)
```

```python
save_folder = 'test'
os.makedirs(save_folder, exist_ok=True)  # Create folder if it doesn't exist

# Save each array as an image
for i, arr in enumerate(all_imgs):
    # Save as PNG (you can choose other formats like .jpg, .bmp, etc.)
    file_path = os.path.join(save_folder, f"image_{i+1}.png")
    plt.imsave(file_path, arr)  # For grayscale images
    print(f"Saved image {i+1} as {file_path}")
```
